## 함수
1. 함수 생성 - def 키워드를 이용해서 생성
2. 매개변수 0개부터 생성
3. 매개변수의 개수와 인자의 개수를 같아야한다. 
4. 매개변수에 기본값을 지정하면 인자에 값을 넣지 않아도 함수는 실행

```python
## 함수 생성
def func_1(a, b, c, d):
    print(a, b, c, d)

## 함수 호출(매개변수의 개수와 인자의 개수를 동일하게)
func_1(1, 2, 3, 4)
func_1(4, 3, 2, 1)
```

```python
## 매개변수에 기본값 지정하여 함수 생성
def func_2(a, b, c=3, d=4):
    print(a, b, c, d)

## 함수 호출
func_2(1, 2)
func_2(1, 2, 5, 6)
```
